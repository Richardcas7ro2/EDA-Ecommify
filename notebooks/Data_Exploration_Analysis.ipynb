```python
#Definición de ruta de archivos
CARPETA_DRIVE = "/content/drive/MyDrive/Proyecto_Ecommify_Olist/datos"

customers = pd.read_csv(f"{CARPETA_DRIVE}/olist_customers_dataset.csv")
geolocation = pd.read_csv(f"{CARPETA_DRIVE}/olist_geolocation_dataset.csv")
order_items = pd.read_csv(f"{CARPETA_DRIVE}/olist_order_items_dataset.csv")
payments = pd.read_csv(f"{CARPETA_DRIVE}/olist_order_payments_dataset.csv")
reviews = pd.read_csv(f"{CARPETA_DRIVE}/olist_order_reviews_dataset.csv")
orders = pd.read_csv(f"{CARPETA_DRIVE}/olist_orders_dataset.csv")
products = pd.read_csv(f"{CARPETA_DRIVE}/olist_products_dataset.csv")
sellers = pd.read_csv(f"{CARPETA_DRIVE}/olist_sellers_dataset.csv")
category_translation = pd.read_csv(f"{CARPETA_DRIVE}/product_category_name_translation.csv")

tablas = {
    "customers": customers,
    "geolocation": geolocation,
    "order_items": order_items,
    "payments": payments,
    "reviews": reviews,
    "orders": orders,
    "products": products,
    "sellers": sellers,
    "category_translation": category_translation
}
```

```python
resumen_nulos = []

for nombre, df in tablas.items():
    total_filas = len(df)
    for columna in df.columns:
        cantidad_nulos = df[columna].isnull().sum()
        porcentaje_nulos = round((cantidad_nulos / total_filas) * 100, 2)

        resumen_nulos.append({
            "tabla": nombre,
            "columna": columna,
            "nulos": cantidad_nulos,
            "porcentaje_nulos": porcentaje_nulos
        })

df_nulos = pd.DataFrame(resumen_nulos)
df_nulos = df_nulos.sort_values(by="porcentaje_nulos", ascending=False)

display(df_nulos)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(
    data=df_nulos[df_nulos["porcentaje_nulos"] > 0],
    x="porcentaje_nulos",
    y="columna",
    hue="tabla"
)
plt.title("Porcentaje de valores nulos por columna")
plt.xlabel("% de nulos")
plt.ylabel("Columna")
plt.show()
```

```python
duplicados = []

for nombre, df in tablas.items():
    duplicados.append({
        "tabla": nombre,
        "filas": len(df),
        "duplicados": df.duplicated().sum()
    })

df_duplicados = pd.DataFrame(duplicados)
display(df_duplicados)
```

```python
validaciones = {
    "orders.customer_id en customers": orders["customer_id"].isin(customers["customer_id"]).mean() * 100,
    "order_items.order_id en orders": order_items["order_id"].isin(orders["order_id"]).mean() * 100,
    "payments.order_id en orders": payments["order_id"].isin(orders["order_id"]).mean() * 100,
    "reviews.order_id en orders": reviews["order_id"].isin(orders["order_id"]).mean() * 100,
    "order_items.product_id en products": order_items["product_id"].isin(products["product_id"]).mean() * 100,
    "order_items.seller_id en sellers": order_items["seller_id"].isin(sellers["seller_id"]).mean() * 100,
    "products.category en translation": products["product_category_name"].isin(category_translation["product_category_name"]).mean() * 100
}

df_validaciones = pd.DataFrame(
    validaciones.items(),
    columns=["validacion", "porcentaje_coincidencia"]
)

display(df_validaciones)
```

```python
columnas_fecha_orders = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in columnas_fecha_orders:
    orders[col] = pd.to_datetime(orders[col], errors="coerce")

order_items["shipping_limit_date"] = pd.to_datetime(order_items["shipping_limit_date"], errors="coerce")
reviews["review_creation_date"] = pd.to_datetime(reviews["review_creation_date"], errors="coerce")
reviews["review_answer_timestamp"] = pd.to_datetime(reviews["review_answer_timestamp"], errors="coerce")
```

```python
df_ecommerce = orders.merge(customers, on="customer_id", how="left")

df_ecommerce = df_ecommerce.merge(order_items, on="order_id", how="left")

df_ecommerce = df_ecommerce.merge(products, on="product_id", how="left")

df_ecommerce = df_ecommerce.merge(sellers, on="seller_id", how="left")

df_ecommerce = df_ecommerce.merge(payments, on="order_id", how="left")

df_ecommerce = df_ecommerce.merge(reviews, on="order_id", how="left")

df_ecommerce = df_ecommerce.merge(category_translation, on="product_category_name", how="left")

print("Dimensiones del dataset consolidado:", df_ecommerce.shape)
display(df_ecommerce.head())
```

```python
df_ecommerce["anio"] = df_ecommerce["order_purchase_timestamp"].dt.year
df_ecommerce["mes"] = df_ecommerce["order_purchase_timestamp"].dt.month
df_ecommerce["anio_mes"] = df_ecommerce["order_purchase_timestamp"].dt.to_period("M").astype(str)
df_ecommerce["dia_semana"] = df_ecommerce["order_purchase_timestamp"].dt.day_name()
```

```python
pedidos_mes = (
    orders
    .assign(anio_mes=orders["order_purchase_timestamp"].dt.to_period("M").astype(str))
    .groupby("anio_mes")["order_id"]
    .nunique()
    .reset_index(name="cantidad_pedidos")
)

plt.figure(figsize=(14,6))
sns.lineplot(data=pedidos_mes, x="anio_mes", y="cantidad_pedidos", marker="o")
plt.xticks(rotation=90)
plt.title("Cantidad de pedidos por mes")
plt.xlabel("Año-Mes")
plt.ylabel("Cantidad de pedidos")
plt.show()
```

```python
clientes_estado = (
    customers
    .groupby("customer_state")["customer_id"]
    .nunique()
    .reset_index(name="cantidad_clientes")
    .sort_values(by="cantidad_clientes", ascending=False)
)

plt.figure(figsize=(12,6))
sns.barplot(data=clientes_estado, x="customer_state", y="cantidad_clientes")
plt.title("Cantidad de clientes por estado")
plt.xlabel("Estado")
plt.ylabel("Cantidad de clientes")
plt.show()

display(clientes_estado)
```

```python
sellers_estado = (
    sellers
    .groupby("seller_state")["seller_id"]
    .nunique()
    .reset_index(name="cantidad_vendedores")
    .sort_values(by="cantidad_vendedores", ascending=False)
)

plt.figure(figsize=(12,6))
sns.barplot(data=sellers_estado, x="seller_state", y="cantidad_vendedores")
plt.title("Cantidad de vendedores por estado")
plt.xlabel("Estado")
plt.ylabel("Cantidad de vendedores")
plt.show()

display(sellers_estado)
```

```python
estado_pedidos = (
    orders["order_status"]
    .value_counts()
    .reset_index()
)

estado_pedidos.columns = ["estado_pedido", "cantidad"]

plt.figure(figsize=(10,5))
sns.barplot(data=estado_pedidos, x="estado_pedido", y="cantidad")
plt.title("Distribución de estados de los pedidos")
plt.xlabel("Estado")
plt.ylabel("Cantidad")
plt.xticks(rotation=45)
plt.show()

display(estado_pedidos)
```

```python
pagos_tipo = (
    payments
    .groupby("payment_type")["payment_value"]
    .agg(["count", "sum", "mean"])
    .reset_index()
    .sort_values(by="sum", ascending=False)
)

display(pagos_tipo)

plt.figure(figsize=(10,5))
sns.barplot(data=pagos_tipo, x="payment_type", y="sum")
plt.title("Valor total pagado por tipo de pago")
plt.xlabel("Tipo de pago")
plt.ylabel("Valor total")
plt.show()
```

```python
categorias = (
    df_ecommerce
    .groupby("product_category_name_english")["order_id"]
    .nunique()
    .reset_index(name="cantidad_pedidos")
    .sort_values(by="cantidad_pedidos", ascending=False)
    .head(15)
)

plt.figure(figsize=(12,7))
sns.barplot(data=categorias, x="cantidad_pedidos", y="product_category_name_english")
plt.title("Top 15 categorías con más pedidos")
plt.xlabel("Cantidad de pedidos")
plt.ylabel("Categoría")
plt.show()

display(categorias)
```

```python
reviews_score = (
    reviews
    .groupby("review_score")["review_id"]
    .count()
    .reset_index(name="cantidad_reviews")
)

plt.figure(figsize=(8,5))
sns.barplot(data=reviews_score, x="review_score", y="cantidad_reviews")
plt.title("Distribución de calificaciones de reseñas")
plt.xlabel("Calificación")
plt.ylabel("Cantidad")
plt.show()

display(reviews_score)
```

```python
metricas_generales = {
    "total_clientes": customers["customer_unique_id"].nunique(),
    "total_pedidos": orders["order_id"].nunique(),
    "total_productos": products["product_id"].nunique(),
    "total_vendedores": sellers["seller_id"].nunique(),
    "total_items_vendidos": len(order_items),
    "valor_total_pagos": payments["payment_value"].sum(),
    "ticket_promedio": payments["payment_value"].mean(),
    "fecha_minima_pedido": orders["order_purchase_timestamp"].min(),
    "fecha_maxima_pedido": orders["order_purchase_timestamp"].max()
}

df_metricas = pd.DataFrame(metricas_generales.items(), columns=["metrica", "valor"])
display(df_metricas)
```
